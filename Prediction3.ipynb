```python
import pandas as pd
import numpy as np
```

## Load data

```python
df = pd.DataFrame()
for i in np.arange(401):
    df = pd.concat([df,pd.read_csv("data/shifts_new/"+str(i)+".csv")],ignore_index=True)
```

```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
```

## Create More Features

```python
df["hour"] = df["Timestamp"].apply(lambda x: x.hour)
```

```python
df["weekday"] = df["Timestamp"].apply(lambda x: x.weekday())
```

```python
df.head()
```

## Split data into training and testing sets

```python
from datetime import datetime, timedelta

bus_period = max(df["Timestamp"])-min(df["Timestamp"])
train = df[df["Timestamp"] < min(df["Timestamp"])+timedelta(days=30*0.8)]
test = df[df["Timestamp"] >= min(df["Timestamp"])+timedelta(days=30*0.8)]
```

## Train a regression model

```python
from sklearn import svm
from sklearn.linear_model import LinearRegression 
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.pipeline import Pipeline


# lm = LinearRegression()
# lm.fit(train[["Lon","Lat","hour","tostop"]], train["timeleft"]) 

rf = RandomForestRegressor()
rf.fit(train[["Lon","Lat","hour","weekday","tostop"]], train["timeleft"])

gdb = GradientBoostingRegressor()
gdb.fit(train[["Lon","Lat","hour","weekday","tostop"]], train["timeleft"])
```

```python
# print(lm.score(test[["Lon","Lat","hour","tostop"]], test["timeleft"]))
print(rf.score(test[["Lon","Lat","hour","weekday","tostop"]], test["timeleft"]))
print(gdb.score(test[["Lon","Lat","hour","weekday","tostop"]], test["timeleft"]))
```

## Predict on the testing set

```python
# lm_predictions = lm.predict(test[["Lon","Lat","hour","tostop"]])
rf_predictions = rf.predict(test[["Lon","Lat","hour","weekday","tostop"]])
gdb_predictions = gdb.predict(test[["Lon","Lat","hour","weekday","tostop"]])
```

## Evaluate the predictions


```python
from sklearn.metrics import mean_squared_error,mean_absolute_error

# print("the error using linear regression is +/- ", round(mean_absolute_error(lm_predictions, test["timeleft"])/60,2), " minutes")
print("the error using random forest is +/- ", round(mean_absolute_error(rf_predictions, test["timeleft"])/60,2), " minutes")
print("the error using gradient boosting is +/- ", round(mean_absolute_error(gdb_predictions, test["timeleft"])/60,2), " minutes")
```

## Residual Analysis

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
%matplotlib inline
import matplotlib as mpl
import seaborn as sns
mpl.rcParams['figure.figsize'] = (10,5)

sns.set(font_scale=2) 
ax = sns.regplot(x=test["timeleft"]/60, y=(gdb_predictions-test["timeleft"])/60, x_bins=150, lowess=True)
ax.axhline(y=0,color="tab:red")
ax.set_title('Error vs. Time to Target Stop (min) - Gradient Boosting')
ax.set_xlabel('Time to Target Stop')
ax.set_ylabel('Error')
```

```python
sns.set(font_scale=2) 
ax = sns.regplot(x=test["timeleft"]/60, y=(rf_predictions-test["timeleft"])/60, x_bins=150, lowess=True)
ax.axhline(y=0,color="tab:red")
ax.set_title('Error vs. Time to Target Stop (min) - Random Forest')
ax.set_xlabel('Time to Target Stop')
ax.set_ylabel('Error')
```
